```python
def get_dataloader(batch_size, image_size, data_dir='processed_celeba_small/'):
    """
    Batch the neural network data using DataLoader
    :param batch_size: The size of each batch; the number of images in a batch
    :param img_size: The square size of the image data (x, y)
    :param data_dir: Directory where image data is located
    :return: DataLoader with batched data
    """
    # Define a transformation function
    transformation = transforms.Compose([transforms.Resize(image_size),
                                      transforms.ToTensor()])
    # Load image data
    images = datasets.ImageFolder(data_dir, transform=transformation)
    # Create a data Loader
    imageLoader = torch.utils.data.DataLoader(images, batch_size=batch_size, shuffle=True)
    
    return imageLoader
```

#### Exercise: Pre-process your image data and scale it to a pixel range of -1 to 1

You need to do a bit of pre-processing; you know that the output of a `tanh` activated generator will contain pixel values in a range from -1 to 1, and so, we need to rescale our training images to a range of -1 to 1. (Right now, they are in a range from 0-1.)

```python
# TODO: Complete the scale function
import torchvision.transforms.functional as TF
def scale(x, feature_range=(-1, 1)):
    ''' Scale takes in an image x and returns that image, scaled
       with a feature_range of pixel values from -1 to 1. 
       This function assumes that the input x is already scaled from 0-1.'''
    # assume x is scaled to (0, 1)
    # scale to feature_range and return scaled x
    # First we need to compute the mean and std for the transformation based on the feature_range
    # if feature_range=(r1,r2) it's easy to prove that mean=-r1/(r2-r1) and std=1/(r2-r1)
    mean_val = -feature_range[0]/(feature_range[1]-feature_range[0])
    std_val  = 1/(feature_range[1]-feature_range[0])
    
    x = (x - mean_val)/std_val
    
    return x
```

# Define the Model

A GAN is comprised of two adversarial networks, a discriminator and a generator.

## Discriminator

Your first task will be to define the discriminator. This is a convolutional classifier like you've built before, only without any maxpooling layers. To deal with this complex data, it's suggested you use a deep network with **normalization**. You are also allowed to create any helper functions that may be useful.

#### Exercise: Complete the Discriminator class
* The inputs to the discriminator are 32x32x3 tensor images
* The output should be a single value that will indicate whether a given image is real or fake


```python
import torch.nn as nn
import torch.nn.functional as F
```

```python
class Discriminator(nn.Module):

    def __init__(self, conv_dim):
        self.conv_dim=conv_dim
        """
        Initialize the Discriminator Module
        :param conv_dim: The depth of the first convolutional layer
        """
        super(Discriminator, self).__init__()
        ## Define layers of a CNN
        # Input 3 channels and output "conv_dim" channels (the convolutin operation is going to
        # reduce he 3 channels of 32x32 image to "conv_dim" channels of 16x16)
        self.conv_0 = nn.Conv2d(3,conv_dim,kernel_size=(4,4),stride=(2,2),padding=(1,1),bias=False)
        # Normalize the values
        self.normalization_0 = nn.BatchNorm2d(conv_dim,eps=0.000001,momentum=0.1)
        # Convolution the "conv_dim" images to 2*"conv_dim". This is going to generate 2*"conv_dim" images of
        # 8x8
        self.conv_1 = nn.Conv2d(conv_dim,2*conv_dim,kernel_size=(4,4),stride=(2,2),padding=1,bias=False)    
        # Normalize
        self.normalization_1 = nn.BatchNorm2d(2*conv_dim,eps=0.000001,momentum=0.1,affine=True)
        # Convolution the 2*"conv_dim" images to 4*"conv_dim". This is going to generate 4*"conv_dim" images of
        # 4x4
        self.conv_2 = nn.Conv2d(2*conv_dim,4*conv_dim,kernel_size=(4,4),stride=(2,2),padding=1,bias=False)    
        # Normalize
        self.normalization_2 = nn.BatchNorm2d(4*conv_dim,eps=0.000001,momentum=0.1,affine=True)
        
        # Now we can convert the elements to a vector of 4x4x4xconv_dim units for the 
        # total connected neural networks
        self.classifier_input = nn.Linear(in_features=4*4*4*conv_dim ,out_features=1,bias=False)
        #self.classifier_hidden1 = nn.Linear(in_features=4*4*conv_dim,out_features=4*conv_dim,bias=False)
        #self.classifier_hidden2 = nn.Linear(in_features=4*conv_dim,out_features=conv_dim,bias=False)
        #self.output = nn.Linear(in_features=conv_dim,out_features=1,bias=False)
        

    def forward(self, x):
        """
        Forward propagation of the neural network
        :param x: The input to the neural network     
        :return: Discriminator logits; the output of the neural network
        """
        # define feedforward behavior
        x = self.conv_0(x)
        x = F.leaky_relu(self.normalization_0(x), 0.02)
        x = self.conv_1(x)
        x = F.leaky_relu(self.normalization_1(x), 0.02)
        x = self.conv_2(x)
        x = F.leaky_relu(self.normalization_2(x), 0.02)
        # Reshape data to input in the full connected NN
        x = x.view(-1, 4*4*4*self.conv_dim)
        # Full Connected NN
        x = self.classifier_input(x) # Remove after test
        #x = F.leaky_relu(self.classifier_input(x))
        #x = F.leaky_relu(self.classifier_hidden1(x))
        #x = F.leaky_relu(self.classifier_hidden2(x))
        #x = self.output(x)
        
        return x

"""
DON'T MODIFY ANYTHING IN THIS CELL THAT IS BELOW THIS LINE
"""
tests.test_discriminator(Discriminator)
```

## Generator

The generator should upsample an input and generate a *new* image of the same size as our training data `32x32x3`. This should be mostly transpose convolutional layers with normalization applied to the outputs.

#### Exercise: Complete the Generator class
* The inputs to the generator are vectors of some length `z_size`
* The output should be a image of shape `32x32x3`

```python
class Generator(nn.Module):
    
    def __init__(self, z_size, conv_dim):
        
        self.conv_dim=conv_dim
        """
        Initialize the Generator Module
        :param z_size: The length of the input latent vector, z
        :param conv_dim: The depth of the inputs to the *last* transpose convolutional layer
        """
        super(Generator, self).__init__()
        # Connect z to a fully connected layer
        self.linear_1 = nn.Linear(in_features=z_size,out_features= 4*4*4*conv_dim,bias=False) # Remove after test
        #self.linear_1 = nn.Linear(in_features=z_size,out_features= 2*conv_dim,bias=False)
        #self.linear_2 = nn.Linear(in_features=2*conv_dim,out_features= 4*conv_dim,bias=False)
        #self.linear_3 = nn.Linear(in_features=4*conv_dim,out_features= 4*4*conv_dim,bias=False)
        #self.linear_4 = nn.Linear(in_features=4*4*conv_dim,out_features= 4*4*4*conv_dim,bias=False)

        # create a sequence of transpose + optimal batch norm layers
        self.t_conv1 = nn.ConvTranspose2d(in_channels= 4*conv_dim,out_channels= 2*conv_dim, kernel_size= 4, 
                                          stride= 2, padding= 1)
        self.normalization_1 = nn.BatchNorm2d(2*conv_dim,eps=0.000001,momentum=0.1,affine=True)
        
        self.t_conv2 = nn.ConvTranspose2d(in_channels= 2*conv_dim,out_channels= conv_dim, kernel_size= 4, 
                                          stride= 2, padding= 1)
        self.normalization_2 = nn.BatchNorm2d(conv_dim,eps=0.000001,momentum=0.1,affine=True)
        
        self.t_conv3 = nn.ConvTranspose2d(in_channels= conv_dim,out_channels= 3, kernel_size= 4, 
                                          stride= 2, padding= 1, bias= False)
        
        

    def forward(self, x):
        """
        Forward propagation of the neural network
        :param x: The input to the neural network     
        :return: A 32x32x3 Tensor image as output
        """
        # define feedforward behavior
        # Fully conected NN
        x = F.relu(self.linear_1(x))
        #x = F.relu(self.linear_2(x))
        #x = F.relu(self.linear_3(x))
        #x = F.relu(self.linear_4(x))
        
        # Reshape x (tensor with size [conv_dim*4,4,4])
        x = x.view(-1,self.conv_dim*4,4,4) 
        
        # Transpose convolution
        x = self.t_conv1(x)
        x = F.relu(self.normalization_1(x))
        
        x = self.t_conv2(x)
        x = F.relu(self.normalization_2(x))
        
        x = self.t_conv3(x)
        x = F.tanh(x)
         
        return x

"""
DON'T MODIFY ANYTHING IN THIS CELL THAT IS BELOW THIS LINE
"""
tests.test_generator(Generator)
```

#### Exercise: Complete the weight initialization function

* This should initialize only **convolutional** and **linear** layers
* Initialize the weights to a normal distribution, centered around 0, with a standard deviation of 0.02.
* The bias terms, if they exist, may be left alone or set to 0.

```python
def weights_init_normal(m):
    """
    Applies initial weights to certain layers in a model .
    The weights are taken from a normal distribution 
    with mean = 0, std dev = 0.02.
    :param m: A module or layer in a network    
    """
    # classname will be something like:
    # `Conv`, `BatchNorm2d`, `Linear`, etc.
    classname = m.__class__.__name__
    # TODO: Apply initial weights to convolutional and linear layers
    if hasattr(m, 'weight') and (classname.find('Conv') != -1 or classname.find('Linear') != -1):
        init.normal_(m.weight.data, 0.0, 0.02)
        if hasattr(m, 'bias') and m.bias is not None:
            init.constant_(m.bias.data, 0.0)
    elif classname.find('BatchNorm2d') != -1:  # BatchNorm Layer's weight is not a matrix; only normal distribution applies.
        init.normal_(m.weight.data, 1.0, 0.02)
        init.constant_(m.bias.data, 0.0)
    
    
```

## Discriminator and Generator Losses

Now we need to calculate the losses for both types of adversarial networks.

### Discriminator Losses

> * For the discriminator, the total loss is the sum of the losses for real and fake images, `d_loss = d_real_loss + d_fake_loss`. 
* Remember that we want the discriminator to output 1 for real images and 0 for fake images, so we need to set up the losses to reflect that.


### Generator Loss

The generator loss will look similar only with flipped labels. The generator's goal is to get the discriminator to *think* its generated images are *real*.

#### Exercise: Complete real and fake loss functions

**You may choose to use either cross entropy or a least squares error loss to complete the following `real_loss` and `fake_loss` functions.**

```python
def real_loss(D_out):
    '''Calculates how close discriminator outputs are to being real.
       param, D_out: discriminator logits
       return: real loss'''
    # Extract the batch size
    batch_size= D_out.size(0)
    # Create the labels under the assumption of Real images (labels=1)
    labels = torch.ones(batch_size)
    # Move to GPU is available
    if train_on_gpu:
        labels = labels.cuda()
    # Loss: Binary Cross Entropy with with Logits loss
    criterion = nn.BCEWithLogitsLoss()
    loss = criterion(D_out.squeeze(), labels)
    return loss
# And now the same but considering that the images is fake
def fake_loss(D_out):
    '''Calculates how close discriminator outputs are to being fake.
       param, D_out: discriminator logits
       return: fake loss'''
    # Extract the batch size
    batch_size= D_out.size(0)
    # Create the labels under the assumption of Fake images (labels=0)
    labels = torch.zeros(batch_size)
    # Move to GPU is available
    if train_on_gpu:
        labels = labels.cuda()
    # Loss: Binary Cross Entropy with with Logits loss
    criterion = nn.BCEWithLogitsLoss()
    loss = criterion(D_out.squeeze(), labels)

    return loss
```

## Training

Training will involve alternating between training the discriminator and the generator. You'll use your functions `real_loss` and `fake_loss` to help you calculate the discriminator losses.

* You should train the discriminator by alternating on real and fake images
* Then the generator, which tries to trick the discriminator and should have an opposing loss function


#### Saving Samples

You've been given some code to print out some loss statistics and save some generated "fake" samples.

#### Exercise: Complete the training function

Keep in mind that, if you've moved your models to GPU, you'll also have to move any model inputs to GPU.

```python
def train(D, G, n_epochs, print_every=50):
    '''Trains adversarial networks for some number of epochs
       param, D: the discriminator network
       param, G: the generator network
       param, n_epochs: number of epochs to train for
       param, print_every: when to print and record the models' losses
       return: D and G losses'''
    
    # move models to GPU
    if train_on_gpu:
        D.cuda()
        G.cuda()

    # keep track of loss and generated, "fake" samples
    samples = []
    losses = []

    # Get some fixed data for sampling. These are images that are held
    # constant throughout training, and allow us to inspect the model's performance
    sample_size=16
    fixed_z = np.random.uniform(-1, 1, size=(sample_size, z_size))
    fixed_z = torch.from_numpy(fixed_z).float()
    # move z to GPU if available
    if train_on_gpu:
        fixed_z = fixed_z.cuda()

    # epoch training loop
    for epoch in range(n_epochs):

        # batch training loop
        for batch_i, (real_images, _) in enumerate(celeba_train_loader):
            
            batch_size = real_images.size(0)
            # Real images
            real_images = scale(real_images)
            # Generate fake images (z)
            z = np.random.uniform(-1, 1, size=(batch_size, z_size))
            z = torch.from_numpy(z).float()

            # ===============================================
            #         YOUR CODE HERE: TRAIN THE NETWORKS
            # ===============================================
            # Initialize to zero the optimizer
            d_optimizer.zero_grad()
            # Send the images to CUDA id available
            if train_on_gpu:
                real_images = real_images.cuda() # Real images
                z = z.cuda() # Fake images
            # Create the final fake image
            fake_images = G(z)
            
            ################ DISCRIMINATOR NN #################
            # Discriminator in real and fake images
            D_real = D(real_images)
            D_fake = D(fake_images)
            
            # Real and fake loss of the discriminator NN
            d_real_loss = real_loss(D_real)
            d_fake_loss = fake_loss(D_fake)
            # Total loss
            d_loss = d_real_loss + d_fake_loss 
            # Back propagation
            d_loss.backward()
            # Optimization
            d_optimizer.step()
            
            ############# END DISCRIMINATOR NN ###############
            
            ################# GENERATOR NN ###################
            # This is almost the same than in above
            # set thegradient to zero
            g_optimizer.zero_grad()
            # Generate fake images (z)
            z = np.random.uniform(-1, 1, size=(batch_size, z_size))
            z = torch.from_numpy(z).float()
            # Send the images to CUDA if available
            if train_on_gpu:
                z = z.cuda() # Fake images
            # Fake images
            fake_images = G(z)
            # Compute the discriminatory in the fake images
            D_fake = D(fake_images)
            # Compute the loss in the generative NN
            g_loss = real_loss(D_fake)
            # Backpropagation
            g_loss.backward()
            # Optimize
            g_optimizer.step()            
            
            # ===============================================
            #              END OF YOUR CODE
            # ===============================================

            # Print some loss stats
            if batch_i % print_every == 0:
                # append discriminator loss and generator loss
                losses.append((d_loss.item(), g_loss.item()))
                # print discriminator and generator loss
                print('Epoch [{:5d}/{:5d}] | d_loss: {:6.4f} | g_loss: {:6.4f}'.format(
                        epoch+1, n_epochs, d_loss.item(), g_loss.item()))


        ## AFTER EACH EPOCH##    
        # this code assumes your generator is named G, feel free to change the name
        # generate and save sample, fake images
        G.eval() # for generating samples
        samples_z = G(fixed_z)
        samples.append(samples_z)
        G.train() # back to training mode

    # Save training generator samples
    with open('train_samples.pkl', 'wb') as f:
        pkl.dump(samples, f)
    # Save the models
    torch.save(G.state_dict(), 'Generation_NN.pt')
    torch.save(D.state_dict(), 'Discriminator_NN.pt')
    # finally return losses
    return losses
```

Of course, intuitively, largest or more complex Generators, probably are going to obtain better results, but in an active field of investigation like this, where a lot of investigations are running to understand this kind of NN, a direct relationship between the complexity of the Generator and the quality of the images is not clear until now.
    * I spent a lot of time tuning the hyperparameters for this model. Initially, I implemented a more deep fully connected network in both, the discriminator, and the generator, but this doesn't seem to improve the results in comparison with a model with a single linear layer.
    * Other option here is to prove the model with differents cost functions, but some researchs had suggested that tuning hyperparameters may ripe a better results than changing the cost function ([see here for details](https://towardsdatascience.com/gan-ways-to-improve-gan-performance-acf37f9f59b)).
    * Like in any other Machine Leaning project, a key parameter to tune is the learning rate. In this case, and after tests the model with different values, I took the path to set different learning rates for the discriminator and generator. The main consideration here is that the discriminator reaches a good accuracy very fast while the generator takes more steps in the process to reach competitive results. Nevertheless, these models are very sensitive and collapse really fast, so, I was conservative with these two learning rates and set it in: *lr_discriminator = 0.0001* and *lr_generator = 0.00025*.
    * I had a lot of work on the number of epochs. In this case, the best visual results were obtained with 7 epochs, while I observed a clear deterioration of the generated images and loss function with epochs around 50.
    
> In my opinion, it's necessary to include more training images with multiple modes (variety) if we want to improve these results. The current generated images look great for this first approach, but in general, need to be improved with strategies similar that the ones described above.
